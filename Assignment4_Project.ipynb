From Wikipedia:
>**Incidence** in epidemiology is a measure of the **probability** of occurrence of a given medical condition in a population within a specified period of time. Although sometimes loosely expressed simply as the number of new cases during some time period, it is better expressed as a proportion or a rate with a denominator.

>**Incidence proportion** (also known as cumulative incidence) is the number of new cases within a specified time period divided by the size of the population initially at risk. For example, if a population initially contains 1,000 non-diseased persons and 28 develop a condition over two years of observation, the incidence proportion is 28 cases per 1,000 persons per two years, i.e. 2.8% per two years.

>*The incidence rate* is the number of new cases per population at risk in a given time period.

>_Incidence should not be confused with prevalence, which is the proportion of cases in the population at a given time rather than rate of occurrence of new cases. Thus, incidence conveys information about the risk of contracting the disease, whereas prevalence indicates how widespread the disease is. Prevalence is the proportion of the total number of cases to the total population and is more a measure of the burden of the disease on society with no regard to time at risk or when subjects may have been exposed to a possible risk factor._

## 1. Data class

```python
class infectData:
    """
    A base class that can be used for retrieving data from Infections Dataset.
    """
    no_top = 15
    
    inc_rate = 1000000
    
    init_county = 'California'
    
    # instantiate the class
    def __init__(self): 
        """
        Initialize the data and a new figure.
        """
        
          # seed for data.
        np.random.seed(12345)
        # get some data to plot

        # read CDC Infections Dataset
        self.df = pd.read_csv('diseases_26.csv')
        self.years = pd.DataFrame(self.df).reset_index()['Year'].unique()[:-1]
        
        # read Poverty US Census Dataset
        poverty_df = pd.read_csv('US_Census_CA-2004_2013_Poverty.csv')\
                        [['Year', 'State / County Name','All Ages in Poverty Percent']]
        poverty_df.columns = ['Year', 'County', 'Poverty Rate']
        poverty_df['County'] = poverty_df['County'].str.replace(' County \(CA\)', '')

        # read Insurance US Census Dataset
        insurance_df = pd.read_csv('US_Census_CA-2006_2013_Insurance.csv')\
                [['Year', 'Name','Uninsured: %']]
        insurance_df.columns = ['Year', 'County', 'Insurance Rate']    
        insurance_df['County'] = insurance_df['County'].str.replace(' County, CA', '')
        
        # write merged Poverty-Insurance Dataset
        self.demogr = pd.merge(poverty_df, insurance_df, on = ['Year', 'County'], how = 'outer').fillna(0)
        
    def set_genders(self):
        self.ap_male = self.get_order_by_year(county = self.init_county,\
                                               sex = 'Male')
        self.ap_female = self.get_order_by_year(county = self.init_county,\
                                               sex = 'Female')
        self.ap_male = pd.melt(self.ap_male.reset_index(), id_vars=['index'], value_vars=self.ap_male.columns)
        self.ap_female = pd.melt(self.ap_female.reset_index(), id_vars=['index'], value_vars=self.ap_female.columns)
        
        self.ap_male["sex"] = 'Male'
        self.ap_female["sex"] = 'Female'
        
    def get_population(self, year, county, sex = 'Total'):
        """
        Returns population in given county at given year with given sex.
        """    
        return self.df[(self.df.Year == year) & (self.df.Sex == sex)\
                  & (self.df.Disease == 'HIV')\
                  & (self.df.County == county)]['Population'].iloc[0]
    
    def get_rate(self, count, year, county, sex = 'Total'):
        """
        Returns disease rate in population in a given county
        at given year with given sex.
        """   
        return self.inc_rate * count / self.get_population(year, county, sex)
    
    def get_count_by_year(self, county = 'all', sex = 'Total'):
        """
        Returns sorted counts of diseases per year.
        Optionally filtered by county and sex.
        """
        if county == 'all':
            ydf = self.df[self.df.Sex == sex].\
                      groupby(['Year', 'Disease']).sum()\
                      ['Count']
        else:
            ydf = self.df[(self.df.Sex == sex) & (self.df.County == county)].\
                      groupby(['Year', 'Disease']).sum()\
                      ['Count']
            
        result_df = pd.DataFrame()
        for yr in self.years[1:]:
            kwargs = {str(yr) : lambda x: ydf[yr].sort_values(ascending = False)}
            result_df = result_df.assign(**kwargs)
        self.count_by_year = result_df
        return self.count_by_year
    
    def get_rate_by_year(self, county = 'all', sex = 'Total'):
        """
        Returns sorted rate of diseases per year.
        Optionally filtered by county and sex.
        
        !!! It COULD be optimized!
        """
        cnt = self.get_count_by_year(county, sex)
        set_rates = lambda x: self.get_rate(x, county, sex)
        
        def set_rates(x):          
            yr_ = int(x.name)
            return x.map(lambda x: self.get_rate(x, yr_, county, sex))

        self.rate_by_year = cnt.apply(set_rates)
        return self.rate_by_year
    
    def get_normalized_rates_year(self, county = 'all', sex = 'Total'):
        """
        Returns 
        
        !!! It COULD be optimized!
        """
        norm_z = lambda x: (x-x.mean())/x.std()
        self.normalized_rates_year = self.get_rate_by_year(county, sex).\
                                        apply(norm_z, axis = 1)
        return self.normalized_rates_year
    
    def scale(self, X, axis=0, with_mean=True, with_std=True):
        """ From scikit-learn

        Standardize a dataset along any axis
        Center to the mean and component wise scale to unit variance.
        Parameters
        ----------
        X : {array-like, sparse matrix}
            The data to center and scale.
        axis : int (0 by default)
            axis used to compute the means and standard deviations along. If 0,
            independently standardize each feature, otherwise (if 1) standardize
            each sample.
        with_mean : boolean, True by default
            If True, center the data before scaling.
        with_std : boolean, True by default
            If True, scale the data to unit variance (or equivalently,
            unit standard deviation).
        """
        
        
        def _handle_zeros_in_scale(scale):
            ''' Makes sure that whenever scale is zero, we handle it correctly.
            This happens in most scalers when we have constant features.'''

            # if we are fitting on 1D arrays, scale might be a scalar
            if np.isscalar(scale):
                if scale == .0:
                    scale = 1.
                return scale
            elif isinstance(scale, np.ndarray):
                scale[scale == 0.0] = 1.0
            return scale
        
        cols, feats, iname = X.columns, X.index.values, X.index.name
        X = np.asarray(X)
        if with_mean:
            mean_ = np.mean(X, axis)
        if with_std:
            scale_ = np.std(X, axis)
        # Xr is a view on the original array that enables easy use of
        # broadcasting on the axis in which we are interested in
        Xr = np.rollaxis(X, axis)
        if with_mean:
            Xr -= mean_
            mean_1 = Xr.mean(axis=0)
            # Verify that mean_1 is 'close to zero'. If X contains very
            # large values, mean_1 can also be very large, due to a lack of
            # precision of mean_. In this case, a pre-scaling of the
            # concerned feature is efficient, for instance by its mean or
            # maximum.
            if not np.allclose(mean_1, 0):
                warnings.warn("Numerical issues were encountered "
                              "when centering the data "
                              "and might not be solved. Dataset may "
                              "contain too large values. You may need "
                              "to prescale your features.")
                Xr -= mean_1
        if with_std:
            scale_ = _handle_zeros_in_scale(scale_)
            Xr /= scale_
            if with_mean:
                mean_2 = Xr.mean(axis=0)
                # If mean_2 is not 'close to zero', it comes from the fact that
                # scale_ is very small so that mean_2 = mean_1/scale_ > 0, even
                # if mean_1 was close to zero. The problem is thus essentially
                # due to the lack of precision of mean_. A solution is then to
                # subtract the mean again:
                if not np.allclose(mean_2, 0):
                    warnings.warn("Numerical issues were encountered "
                                  "when scaling the data "
                                  "and might not be solved. The standard "
                                  "deviation of the data is probably "
                                  "very close to 0. ")
                    Xr -= mean_2
        X = pd.DataFrame(X, columns = cols)
        X[iname] = feats
        return X.set_index(iname)
    
    def get_order_by_year(self, county = 'all', sex = 'Total'):
        """
        Returns list of Diseases ranges every year based on sorted Rate
        
        !!! It COULD be optimized!
        """
        
        self.order_by_year = self.get_rate_by_year(county, sex).\
                                     apply(lambda x:\
                                     pd.Series(np.arange(1,len(x)+1,1),\
                                     index=x.sort_values(ascending = False).index,\
                                     name=x.name))
        return self.order_by_year
        
    def get_order_by_year_buf(self):
        if hasattr(self, 'order_by_year'):
            return self.order_by_year
    
    def get_normalized_rates_year_buf(self):
        if hasattr(self, 'normalized_rates_year'):
            return self.normalized_rates_year
        
    def get_rate_by_year_buf(self):
        if hasattr(self, 'rate_by_year'):
            return self.rate_by_year
        
    
    def get_count_by_year_buf(self):
        if hasattr(self, 'count_by_year'):
            return self.count_by_year   
        
    
```
